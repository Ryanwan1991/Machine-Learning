```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
'''
filename = str(input('Please input the filename'))
df = pd.read_excel(filename)
n = int(input('Please input number of datas'))
'''
df = pd.read_csv('AR 0.5% variation analysis.csv')
```

```python
plt.figure(figsize=(16,4))
plt.style.use('seaborn-darkgrid')
x = df.iloc[:,0]
y = []
n = 500
for i in range(0,n):
    y.append(df.iloc[:,i])

for i in range(0,n):
    plt.plot(x,y[i],'g-')

#plt.legend()
plt.suptitle('AR Coating', fontsize=16, fontweight='bold')
plt.subplots_adjust(left=0.2, wspace=0.8, top=0.8)
plt.xlabel('wavelength')
plt.ylabel('Reflectance')
plt.ylim((0, 5))
plt.xlim((400, 1050))
```

```python
sns.lineplot(x='wavelength',y='R%',data=df )
```

```python
rs = np.random.RandomState(365)
values = rs.randn(365, 4).cumsum(axis=0)
dates = pd.date_range("1 1 2016", periods=365, freq="D")
data = pd.DataFrame(values, dates, columns=["A", "B", "C", "D"])
data = data.rolling(7).mean()

sns.lineplot(data=data, palette="tab10", linewidth=2.5)
```
